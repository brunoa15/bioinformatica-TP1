## Algoritmo de Needleman-Wunsch com penalidade

```python
def needleman_wunsch(seq_a, seq_b):
    seq_a_alinhada = ''
    seq_b_alinhada = ''
    n_linhas = len(seq_b) + 1
    n_colunas = len(seq_a) + 1
    matriz_numerica = np.zeros((n_linhas, n_colunas), dtype=int)
    matriz_resultante = np.zeros((n_linhas, n_colunas), dtype='<U10')
    
    for i in range(n_linhas):
        for j in range(n_colunas):
            
            if i == 0 or j == 0:
                matriz_resultante[i,j] = '0'
                continue
                
            if seq_b[i-1] == seq_a[j-1] or matriz_blosum.loc[seq_a[j-1], seq_b[i-1]] >= 0:
                matriz_numerica[i,j] = matriz_numerica[i-1,j-1] + matriz_blosum.loc[seq_a[j-1], seq_b[i-1]]
                matriz_resultante[i,j] = np.str(matriz_numerica[i,j]) + '\\'
                
            elif matriz_numerica[i,j-1] > matriz_numerica[i-1,j]:
                matriz_numerica[i,j] = matriz_numerica[i,j-1] + matriz_blosum.loc[seq_a[j-1], seq_b[i-1]]
                matriz_resultante[i,j] = np.str(matriz_numerica[i,j]) + '_'
                
            else:
                matriz_numerica[i,j] = matriz_numerica[i-1,j] + matriz_blosum.loc[seq_a[j-1], seq_b[i-1]]
                matriz_resultante[i,j] = np.str(matriz_numerica[i,j]) + '|'
            
    resultado = pd.DataFrame(data=matriz_resultante, columns=list('*' + seq_a))
    resultado[''] = list('*' + seq_b)
    resultado.set_index('', inplace=True)
    
    while i != 0 or j != 0:
        if resultado.iloc[i,j].count('\\'):
            seq_a_alinhada = seq_a_alinhada + resultado.columns[j]
            seq_b_alinhada = seq_b_alinhada + resultado.index[i]
            i = i-1
            j = j-1
        if resultado.iloc[i,j].count('_') or (i == 0 and j > 0):
            seq_a_alinhada = seq_a_alinhada + resultado.columns[j]
            seq_b_alinhada = seq_b_alinhada + '-'
            j = j-1
        if resultado.iloc[i,j].count('|') or (j == 0 and i > 0):
            seq_a_alinhada = seq_a_alinhada + '-'
            seq_b_alinhada = seq_b_alinhada + resultado.index[i]
            i = i-1
    
    return resultado, seq_a_alinhada[::-1], seq_b_alinhada[::-1]
```

## Implementação da visualização dos resultados dos alinhamentos utilizando a biblioteca plotly

O código abaixo utiliza a biblioteca Plotly para gerar um Heatmap que permita a visualização gráfica do resultado dos alinhamentos. O que foi proposto na utilização desse gráfico é mostrar em azul o casamento de
resíduos de aminoácidos idênticos e em vermelho os indels.




---


[Heatmaps in Python](https://plotly.com/python/heatmaps/)



```python
def heatMap(seq_a_alinhado, seq_b_alinhado):
  datas = []
  datas.append(seq_a_alinhado)
  datas.append(seq_b_alinhado)

  dataset = []
  x = []
  y = []
  for i, seq in enumerate(datas):
    y.append(seq)
    value = []
    am = []
    for j, letter in enumerate(seq):
      am.append(letter)
      if (letter == '-'):
        value.append(0)
      else:
        value.append(1)
    dataset.append(value)
    x.append(am)

  fig = go.Figure(data=go.Heatmap(
                    z=dataset,
                    y=['Sq_A', 'Sq_B'],
                    colorscale=[
                      # 
                      [0, "rgb(165,0,38)"],
                      [1.0, "rgb(49,54,149)"]]
                    ))
  fig.show()
```

Além da implemntação da visualização utilizando o Plotfy o grupo utilizou a biblioteca Dash Bio para gerar um visuzalização mais completa de todas as sequencias. A utilização dessa visualização foi importante para percebermos de forma mais clara onde houve mutações, matches, mismatches e indels.

```python
%%writefile visualizacao_dash_app.py

import dash
import dash_bio as dashbio
import dash_html_components as html

import six.moves.urllib.request as urlreq
from six import PY3

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app = dash.Dash(__name__, external_stylesheets=external_stylesheets)

# Sequências dos vírus

data = urlreq.urlopen(
    'https://raw.githubusercontent.com/brunoa15/bioinformatica-TP1/main/sequences.txt'
).read()

if PY3:
    data = data.decode('utf-8')

app.layout = html.Div([
    dashbio.AlignmentChart(
        id='my-alignment-viewer',
        data=data
    ),
    html.Div(id='alignment-viewer-output')
])


@app.callback(
    dash.dependencies.Output('alignment-viewer-output', 'children'),
    [dash.dependencies.Input('my-alignment-viewer', 'eventDatum')]
)
def update_output(value):
    if value is None:
        return 'No data.'
    return str(value)


if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import itertools
import numpy as np

def matriz_alinhamento(a, b, match_score=3, gap_cost=2):
    # calucular o escore maximo entre duas sequencias de caracteres
    # matriz de pontuação
    H = np.zeros((len(a) + 1, len(b) + 1), np.int)
    for i, j in itertools.product(range(1, H.shape[0]), range(1, H.shape[1])):
        match = H[i - 1, j - 1] + (match_score if a[i - 1] == b[j - 1] else - match_score)
        delete = H[i - 1, j] - gap_cost
        insert = H[i, j - 1] - gap_cost
        H[i, j] = max(match, delete, insert, 0)
    return H

def traceback(H, b, b_='', old_i=0):
    # Após obter o H (score maximo), fazemos o  caminho de volta de quias elementos aquele valor se originou
    H_flip = np.flip(np.flip(H, 0), 1)
    i_, j_ = np.unravel_index(H_flip.argmax(), H_flip.shape)
    # (i, j) é a utilma posicao do maior escore
    i, j = np.subtract(H.shape, (i_ + 1, j_ + 1))
    if H[i, j] == 0:
        return b_, j
    b_ = b[j - 1] + '-' + b_ if old_i - i > 1 else b[j - 1] + b_
    return traceback(H[0:i, 0:j], b, b_, i)

def smith_waterman(a, b, match_score=3, gap_cost=2):
    a, b = a.upper(), b.upper()
    H = matrix(a, b, match_score, gap_cost)
    b_, pos = traceback(H, b)
    return pos, pos + len(b_)
```
